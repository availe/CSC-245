```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import pandas as pd
```

```python
data = pd.read_csv('src/company.csv')
```

Exercise 1: Read Total profit of all months and show it using a line plot.

```python
plt.plot(data['month_number'],data['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit per Month')
```

Exercise 2: Get total profit of all months and show line plot with the following Style properties.

```python
plt.plot(data['month_number'],data['total_profit'], linestyle='dotted', color='red', marker='o', linewidth=3, label='Profit data of last year')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit per Month')
plt.legend(loc='lower right')
```

Exercise 3: Read all product sales data and show it using a multiline plot.

```python
products = data.columns[:7]
for product in products:
    plt.plot(data['month_number'],data[product], marker='o', label=product)
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Sales Data')
plt.legend(loc='upper left')
```

Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot.

```python
plt.scatter(data['month_number'], data['toothpaste'], label='Toothpaste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Toothpaste Sales Data')
plt.legend(loc='upper left')
plt.grid(True)
```

Exercise 5: Read face cream and face wash product sales data and show it using the bar chart.

```python
width = 0.25

plt.bar(data['month_number'] - width / 2, data['facewash'], label='Face Wash Sales Data', color='blue', width=width)
plt.bar(data['month_number'] + width / 2, data['facecream'], label='Face Cream Sales Data', color='orange', width=width)
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Facial Products Sales Data')
plt.legend(loc='upper left')
```

Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk.

```python
plt.bar(data['month_number'],data['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Bathing Soap Sales Data')
```

Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges.

```python
bins = [x * 10**4 for x in [15, 17.5, 20, 22.5, 25, 30, 35]]
plt.hist(data['total_profit'], bins=bins, edgecolor='black', rwidth=0.8)
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
```

Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart.

```python
sale_sums = data.iloc[:,:7].sum()
plt.pie(sale_sums, wedgeprops={'edgecolor': 'black'}, autopct='%1.2f%%')
```
