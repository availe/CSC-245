```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import pandas as pd
```

```python
data = pd.read_csv('src/company.csv')
```

Exercise 1: Read Total profit of all months and show it using a line plot

```python
plt.plot(data['month_number'],data['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit per Month')
plt.show()
```

Exercise 2: Get total profit of all months and show line plot with the following Style properties

```python
plt.plot(data['month_number'],data['total_profit'], linestyle='dotted', color='red', marker='o', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit per Month')
plt.legend('Total Profit', loc='lower right')
plt.show()
```
